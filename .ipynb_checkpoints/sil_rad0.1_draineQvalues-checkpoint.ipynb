```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import ascii
```

```python
tabl = ascii.read("draine_silicate0.1")
```

```python
plt.xlabel("Wavelengths (microns)", size = 12)
plt.ylabel("Q", size = 12)
plt.plot(tabl["w(micron)"], tabl["Q_sca"] + tabl["Q_abs"], "b", label = "Q_ext")
plt.plot(tabl["w(micron)"], tabl["Q_sca"], "g", label = "Q_sca")
plt.plot(tabl["w(micron)"],tabl["Q_abs"], "r", label = "Q_abs")
plt.legend(loc = 'lower left')
plt.loglog()

plt.show()
```
